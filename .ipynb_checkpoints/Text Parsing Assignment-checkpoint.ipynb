# Romeo and Juliet Text Parsing
This code is looking for a text file in the same folder as the ipython notebook. The .read().split() command reads in the text file and then splits the text file into discrete words. A counter is initialized as well as a list of words that indicate sorrow/sadness. The code then goes through all of the words in Romeo and Juliet and adds 1 to the counter every time that one of the words was found. 

```python
import random
words = open('RomeoandJuliet.txt').read().split()
counter = 0
words_of_sorrow = ['sad','upset', 'depressed','sorrow','weep','trajedy','regret','misery']

for i in words:
    if i in words_of_sorrow:
        counter += 1
        
print("The total number of words of sorrow in Romeo and Juliet was", counter)

#Finds the percentage
print((counter/len(words))*100)

    
#Prints a random word from the play

random.choice(words)
```

# Punctuation Removal
The .split() function splits the text file into a list using the space character to separate the file. When there is punctuation present in the file this can cause a word to not be recognized by the counter (for instance "Juliet." will not register a hit for the key word "Juliet". To get around this we can remove the punctuation from the text file, split the text file into words, and then search for key words. Notice that this picks up 6 additional words of sorrow that were not found in the previous search.

```python
import random
textfile = open('RomeoandJuliet.txt').read()
punctuations = '''!()-[]{};:'"\,<>./?@#%%ˆ&*˜'''


# Removes the punctuation from the text file
no_punct = ""
for i in textfile:
    if i not in punctuations:
        no_punct = no_punct + i
        
# Splits the punctuation free text into words
words2 = no_punct.split()
counter = 0
words_of_sorrow = ['sad','upset', 'depressed','sorrow','weep','trajedy','regret','misery']

for i in words2:
    if i in words_of_sorrow:
        counter += 1
        
print("The total number of words of sorrow in Romeo and Juliet was", counter)

#Finds the percentage
print((counter/len(words2))*100)

```

# Finding the Most Common Words
We can take use of the library Collections to display the most common words in a list as well as the individual word count. This can be useful to discover key words in a text. 

```python
from collections import Counter

words = open('RomeoandJuliet.txt').read().split()
cnt = Counter(words)
print(cnt.most_common())
```
